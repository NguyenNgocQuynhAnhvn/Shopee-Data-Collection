```python
from pymongo import MongoClient

# Káº¿t ná»‘i Ä‘áº¿n MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["shopee"]  # Thay tháº¿ báº±ng tÃªn database cá»§a báº¡n
collection_shopee = db["collection_shopee"]  # Thay tháº¿ báº±ng tÃªn collection cá»§a báº¡n
collection_feedback = db["feedbacks"]


```

```python
#Truy váº¥n nhá»¯ng feedback cÃ³ lÆ°á»£t Ä‘Ã¡nh giÃ¡ 4 sao
import ast

# Láº¥y tÃ i liá»‡u tá»« collection
rss = collection_shopee.find()  # Chá»‰ láº¥y 2 tÃ i liá»‡u

feedbacks_with_rating_4 = []


for rs in rss:
    feedbacks_list = ast.literal_eval(rs["feedbacks"])  # Äáº£m báº£o ráº±ng 'feedbacks' lÃ  má»™t chuá»—i

    # Lá»c feedbacks cÃ³ star_rating lÃ  4
    feedbacks_with_rating_4.extend(
        [feedback for feedback in feedbacks_list if feedback['star_rating'] == 4]
    )

# # # In ra káº¿t quáº£
print(feedbacks_with_rating_4)
```

```python
#TÃ¬m sá»‘ lÆ°á»£ng pháº©m cÃ³ lÆ°á»£t Ä‘Ã¡nh giÃ¡ lá»›n hÆ¡n 4.8
count = collection_shopee.count_documents({"rate": {"$gt": 4.8}})
print("Sá»‘ lÆ°á»£ng sáº£n pháº©m cÃ³ Ä‘Ã¡nh giÃ¡ > 4.8:", count)
```

```python
#TÃ¬m tÃªn cáº£ sáº£n pháº©m Ä‘Ã£ thu tháº­p Ä‘Æ°á»£c
product_names = collection_shopee.find({}, {"name_product": 1}).limit(5)
print("TÃªn cÃ¡c sáº£n pháº©m:")
for product in product_names:
    if "name_product" in product:
        print(product["name_product"])
    else:
        print("KhÃ´ng cÃ³ khÃ³a 'name_product' trong sáº£n pháº©m")

```

```python
#Kiá»ƒm tra dá»¯ liá»‡u bá»‹ trÃ¹ng láº·p dá»±a vÃ o tÃªn sáº£n pháº©m
pipeline = [
    {
        "$group": {
            "_id": {"name_product": "$name_product"},
            "count": {"$sum": 1},
            "docs": {"$push": "$_id"}
        }
    },
    {
        "$match": {
            "count": {"$gt": 1}
        }
    }
]
# Thá»±c hiá»‡n truy váº¥n
duplicates = list(collection_shopee.aggregate(pipeline))
# In cÃ¡c báº£n ghi trÃ¹ng láº·p
for record in duplicates:
    print(f"Sáº£n pháº©m trÃ¹ng láº·p: {record['_id']}, Sá»‘ láº§n xuáº¥t hiá»‡n: {record['count']}")
    print("IDs cá»§a cÃ¡c báº£n ghi trÃ¹ng láº·p:", record["docs"])
```

```python
#Truy váº¥n thÃ´ng tin sáº£n pháº©m cÃ³ giÃ¡ lá»›n hÆ¡n hoáº·c báº±ng 50.000 vÃ  nhá» hÆ¡n hoáº·c báº±ng 100.000
results = collection_shopee.find({"price_origin": {"$gte": 50000, "$lte": 100000}}).limit(5)
for product in results:
    print(product)
```
