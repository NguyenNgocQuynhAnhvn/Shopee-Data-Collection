```python
from pymongo import MongoClient

# Kết nối đến MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["shopee"]  # Thay thế bằng tên database của bạn
collection_shopee = db["collection_shopee"]  # Thay thế bằng tên collection của bạn
collection_feedback = db["feedbacks"]


```

```python
#Truy vấn những feedback có lượt đánh giá 4 sao
import ast

# Lấy tài liệu từ collection
rss = collection_shopee.find()  # Chỉ lấy 2 tài liệu

feedbacks_with_rating_4 = []


for rs in rss:
    feedbacks_list = ast.literal_eval(rs["feedbacks"])  # Đảm bảo rằng 'feedbacks' là một chuỗi

    # Lọc feedbacks có star_rating là 4
    feedbacks_with_rating_4.extend(
        [feedback for feedback in feedbacks_list if feedback['star_rating'] == 4]
    )

# # # In ra kết quả
print(feedbacks_with_rating_4)
```

```python
#Tìm tất cả các sản phẩm từ shop Shopgiare1234
results = collection_shopee.find({"shop_name": "Shopgiare1234"})
for product in results:
    print(product)
```

```python
#Tìm số lượng phẩm có lượt đánh giá lớn hơn 4.8
count = collection_shopee.count_documents({"rate": {"$gt": 4.8}})
print("Số lượng sản phẩm có đánh giá > 4.8:", count)
```

```python
#Tìm tên cả sản phẩm đã thu thập được
product_names = collection_shopee.find({}, {"name_product": 1}).limit(5)
print("Tên các sản phẩm:")
for product in product_names:
    if "name_product" in product:
        print(product["name_product"])
    else:
        print("Không có khóa 'name_product' trong sản phẩm")

```

```python
#Kiểm tra dữ liệu bị trùng lặp dựa vào tên sản phẩm
pipeline = [
    {
        "$group": {
            "_id": {"name_product": "$name_product"},
            "count": {"$sum": 1},
            "docs": {"$push": "$_id"}
        }
    },
    {
        "$match": {
            "count": {"$gt": 1}
        }
    }
]
# Thực hiện truy vấn
duplicates = list(collection_shopee.aggregate(pipeline))
# In các bản ghi trùng lặp
for record in duplicates:
    print(f"Sản phẩm trùng lặp: {record['_id']}, Số lần xuất hiện: {record['count']}")
    print("IDs của các bản ghi trùng lặp:", record["docs"])
```

```python
#Truy vấn thông tin sản phẩm có giá lớn hơn hoặc bằng 50.000 và nhỏ hơn hoặc bằng 100.000
results = collection_shopee.find({"price_origin": {"$gte": 50000, "$lte": 100000}}).limit(5)
for product in results:
    print(product)
```
